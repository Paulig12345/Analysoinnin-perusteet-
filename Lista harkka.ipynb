```python
number1 = 10
number2 = 5

if number1 == number2:
    print("1a: number1 and number2 are equal")

if number1 > number2:
    print("1b: number1 is greater than number2")

if number1 >= number2:
    print("1c: number1 is greater or equal to number2")

if number1 != number2:
    print("1d: number1 and number2 are NOT equal")
```

```python

number1 = 5
number2 = 5
number3 = 3

if number1 == number2 == number3:
    print("2a: all numbers are equal")

if number1 == number2 or number2 == number3:
    print("2b: number1==number2 OR number2==number3")

if number1 > number2 and number1 > number3:
    print("2c: number1 > number2 AND number1 > number3")

if number1 > number2:
    print("2d: number1 is greater than number2")
else:
    if number2 > number3:
        print("2d: number2 is greater than number3")

if number1 == number2:
    print("2e: number1 and number2 are equal")
else:
    if number1 == number3:
        print("2e: number1 and number3 are equal")
```
