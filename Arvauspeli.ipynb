```python
oikea_nimi = "Pauli"
yritykset = 0

while True:
    arvaus = input("Arvaa nimi: ")
    yritykset += 1

    if arvaus == oikea_nimi:
        print("Onnittelut!")
        print("Arvauksia yhteensä:", yritykset)
        break
    else:
        lopeta = input("Väärin! Haluatko lopettaa? (k/e): ")
        if lopeta.lower() == "k":
            print("Peli päättyi. Arvauksia tehty:", yritykset)
            break
```
